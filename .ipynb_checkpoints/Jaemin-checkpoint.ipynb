```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def dist(a):
    target = [10, 0]
    x_dist = a[0] - target[0]
    y_dist = a[1] - target[1]
    return 10 - math.sqrt(x_dist**2 + y_dist**2)

def sigmoid(x):
    return 1/(1 + math.exp(x))

x1 = [0.01*i for i in range(-100,101)]
x2 = [0.01*i for i in range(100,-101,-1)]
x = x1 + x2
y1 = [math.sqrt(1-x[i]**2) for i in range(101)]
y2 = [math.sqrt(1-x[i]**2) for i in range(101,201)]
y3 = [-math.sqrt(1-x[i]**2) for i in range(101)]
y4 = [-math.sqrt(1-x[i]**2) for i in range(101,201)]
y = y1 + y2 + y3 + y4
location = list(zip(x,y))

reward = list(map(dist, location))
reward = list(map(sigmoid, reward))

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot(x,y,reward)
plt.show()
```
